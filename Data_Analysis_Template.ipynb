```python
import pandas as pd
import datetime
import numpy as np
```

```python
#C:/Users/jiaojiaozhe/Desktop/PWS HW1.csv
df = pd.read_csv ('', header = None, usecols=[0, 1, 2, 3, 7],
                 names=['When','ID','Status','Problem','Author'], converters={'Problem': lambda x: str(x)})
df['When'] = pd.to_datetime(df['When'], format='%Y-%m-%d %H:%M:%S')
pd.set_option('display.max_rows', 100)
```

```python
df.sort_values(['Problem', 'Author'], inplace = True)
df.head(20)
#整理過的資料
```

```python
authors = df['Author'].unique()
authors.sort()
#學生id
problems_id = df['Problem'].unique()
problems_id.sort()
#問題id
len(authors)
```

```python
problem_types = df.groupby(['Author', 'Problem', 'Status'])['Status'].size().rename('Status_Total').reset_index()
problem_types
#分項繳交次數
```

```python
problem_total = df.groupby(['Author', 'Problem'])['Status'].size().rename('Total_Submit_Times').reset_index()
problem_total
#總繳交次數
```

```python
problems_info = pd.merge(problem_types, problem_total, left_on = ['Author', 'Problem'], right_on = ['Author', 'Problem'])
problems_info
#上面兩個合併
```

```python
accepted = problems_info[problems_info["Status"] == "Accepted"]
accepted.sort_values('Status_Total')
```

```python
accepted.insert(5, 'Submit_Times_Before_AC', accepted["Total_Submit_Times"] - accepted["Status_Total"])
times_before_ac = accepted[['Author', 'Problem', 'Submit_Times_Before_AC']]
times_before_ac
```

```python
pd.set_option('display.max_rows', 100)
from_start_to_last_submit = df[['Author', 'Problem', 'When', 'Status']].sort_values(['Author', 'When'])
#from_start_to_last_submit.head(50)
#算時間的表格
```

```python
work_time = []
for author in authors:
    for problem in problems_id:
        submit_time = from_start_to_last_submit.loc[(from_start_to_last_submit['Author'] == author) & (from_start_to_last_submit['Problem'] == problem), ['Author', 'Problem', 'When', 'Status']]
        process_time = submit_time['When'].max() - submit_time['When'].min()
        #print(author, problem, process_time)
        work_time.append([author, problem, process_time])
        
work_time.sort(key=lambda x: (x[0], x[1]))
work_time = pd.DataFrame(work_time, columns= ['Author','Problem', 'Process_Time'])
#work_time
#每人分項所花時間
```

```python
time = pd.merge(times_before_ac, work_time, left_on = ['Author', 'Problem'], right_on = ['Author', 'Problem'])
time.sort_values('Process_Time').head(100)
#合併繳交次數與花的時間
```
